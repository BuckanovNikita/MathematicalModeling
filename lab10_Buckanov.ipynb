## Исследование стационарной модели распределения властных полномочий.

# Идеализация объекта исследования.

$p(i,t)$ - реализуемые полномочия власти

$p_1(i,t)$  - минимальные полномочия власти
 
$p_2(i,t)$  - максимальные полномочия власти $p_2(i+1, t) < p_2(i,t) < p_2(i-1,t)$

$F_s(i,t,p,p_1,p_2)$ - сила рекции общества

$F_h(i,t,p,p_1,p_2)$ - сила рекции иерархии

$W(i,t)$ - поток власти описывающий близкодействие

$V(i+1, j, t)$ - поток власти описывающий дальнодействие

$\frac{\Delta p(i,t)}{\Delta t} = -[W(i,t) - W(i-1,t)] + \sum_{j=0}^{N}V(i,j,t) + F$

граничные условия

$W(-1,t) = W(N,t) = 0$

$p(i,0) = p_0(i)$

# Переход к непрерывной модели.

$\frac{\partial p}{\partial t} = \frac{\partial}{\partial x}
[a_1(p,\frac{\partial p}{\partial x})\frac{\partial p}{\partial x}]
+ \int_0^l a_2(p(x'),p(x),x',x)[p(x',t)-p(x,t)]dx' + F(p,p1,p2,x,t)$

$W(0,t) = -a_2(x)\frac{\partial p}{\partial x}=0$ x = 0

$W(l,t) = -a_1(x)\frac{\partial p}{\partial x}=0$ x = l

$p(x,t_0) = p_0(x)$

# Стационарный случай. Получение аналитического решения. Его анализ в зависимости от параметров системы.

$a_1=const>0$

$a_2=0$

$F(p, x) = k_1(p^0(x)-p)$

$p^0 = H-kx$

$p_1 = (1-\alpha)p^0$

$p_2=(1+\alpha)p^0$

```python
%matplotlib notebook
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import ipywidgets as widgets
import scipy.integrate as integrate
style = {'description_width': 'initial'}
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()

def hierarchy_plotter(a,b):
    ax.clear()
    ax.grid()
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    x = np.linspace(0,1,1000)
    a__ = a**-0.5
    y = a**0.5*b / (np.exp(a__)-np.exp(-a__)) *\
    ((1-np.exp(-a__))*np.exp(x*a__) + (1-np.exp(a__))*np.exp(-x*a__)) + (1-b*x)
    plt.plot(x,y)
    plt.plot(x, 1.8*(-b*x+1))
    plt.plot(x, 0.2*(-b*x+1))
    plt.legend(['Реализуемое', 'Максимальное', 'Минимальное'])
    
interact(hierarchy_plotter,
    a = widgets.FloatSlider(min=0.001, max=0.4, step=0.001, value=0.5),
    b = widgets.FloatSlider(min=0, max=1, step=0.01, value=0.5),
);
```
