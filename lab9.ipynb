## Исследование эффекта снижения и стабилизации цен на потребительские товары доступные среднему классу в экономической модели "Средний класс".

Упрощающие предположения
* Распределение граждан по доходам для среднего класса подчиняется нормальному закону.

* В зоне реальных доходов будет гипотетический товар с некоторой ценой, и в Δ-окрестности от этой цены будет 100% потребления данного товара.


```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()
ind = 0

def plotter(mean_sallary, std_dev, real_price, percent):
    global ind
    if ind%5 == 0:
        ax.clear()
        ax.grid()
    x = np.linspace(real_price, 7000, 10000)
    ax.plot(x, (1 - norm.cdf(100*x/percent, mean_sallary, std_dev))*(x - real_price))
    ind+=1
    
interact(plotter,
    mean_sallary=widgets.IntSlider(min=1000, max=200000, step=1000, value=22000,style = style),
    std_dev=widgets.IntSlider(min=1000, max=10000, step=100, value=5000,style = style),
    real_price = widgets.IntSlider(min=100, max=2000, step=50, value=500,style = style),
    percent = widgets.IntSlider(min=1, max=100, step=1, value=10,style = style)
);
```
