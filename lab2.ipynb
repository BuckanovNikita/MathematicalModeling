Моделирование скорости полета ракеты в зависимости от количества используемых ступеней

```python
import numpy as np
from scipy.constants import g
```

```python
mp = 1
u0 = 10
mass = np.array([10,40,105])
lambda_ = 0.1
m0 = sum(mass)+mp
alpha = np.array([(sum(mass[i:])+mp)/(sum(mass[i+1:])+mp) 
                  for i in range(len(mass))])  
```

```python
un_u0 = np.log([np.product(alpha[:i]/(1+lambda_*(alpha[:i]-1))) 
               for i in range(1,len(alpha)+1)])
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
plt.figure()
plt.plot(un_u0*u0)
plt.title("до")
plt.xlabel("число ступеней")
plt.ylabel("скорость м/с")
plt.grid()
```
