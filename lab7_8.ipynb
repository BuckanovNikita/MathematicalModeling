# моделирование боевых действий

```python
import numpy as np
from scipy.constants import g, pi
import matplotlib.pyplot as plt
from scipy.optimize import minimize
import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed, interact_manual
from scipy.integrate import solve_ivp
```

```python
%matplotlib notebook
style = {'description_width': 'initial'}
```

## $ \frac{d N_1}{d t} = -\alpha_1(t) N_1 - \beta_2(t)  N_2 + \gamma_1(t)$

## $ \frac{d N_2}{d t} = -\alpha_2(t) N_2 - \beta_1(t) N_1 + \gamma_2(t)$

## $ \alpha_1, \alpha_2 $ - потери не связанные с боевыми действиями

## $\beta_1, \beta_2 $ - потери в боевых действиях

## $\gamma_1, \gamma_2$ - скорость получения подкрепления

## простейшая модель случай регулярных армий

```python
def regular_war_model(t, y, a_1, a_2, b_1, b_2, g_1, g_2):
    return np.array([-a_1*y[0] - b_2*y[1] + g_1, -a_2*y[1] - b_1*y[0] + g_2])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()
ax.legend(['Армия 1', 'Армия 2'])

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.grid()
plt.xlabel('Страна 1')


def regular_war_plotter(a_1, a_2, b_1, b_2, g_1, g_2, max_time, start_value_N, start_value_M):
    ax.clear()
    ax1.clear()
    ax1.grid()
    ax.grid()
    plt.xlabel('Армия 1')
    plt.ylabel('Армия 2')

    solve = solve_ivp(lambda t, y: regular_war_model(t, y, a_1, a_2, b_1, b_2, g_1, g_2),
                      (0,max_time), [start_value_N, start_value_M], method = 'RK23')
    
    l1, = ax.plot(solve['t'][(solve['y'][0]>-1) & (solve['y'][1]>-1)], solve['y'][0][(solve['y'][0]>-1) & (solve['y'][1]>-1)])
    l2, = ax.plot(solve['t'][(solve['y'][0]>-1) & (solve['y'][1]>-1)], solve['y'][1][(solve['y'][0]>-1) & (solve['y'][1]>-1)])
    ax1.plot(solve['y'][0][(solve['y'][0]>-1) & (solve['y'][1]>-1)], solve['y'][1][(solve['y'][0]>-1) & (solve['y'][1]>-1)])
    ax.legend((l1, l2), ('Армия 1', 'Армия 2'))
    
interact(regular_war_plotter,
    a_1=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    a_2=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    b_1=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    b_2=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    g_1=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    g_2=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    max_time=widgets.IntSlider(min=4, max=400, step=1, value=5),
    start_value_N=widgets.IntSlider(min=2, max=200, step=10, value=10,style = style),
    start_value_M=widgets.IntSlider(min=2, max=200, step=10, value=10,style = style)
);
```

## простейшая модель случай регулярная армия против партизан

```python
def partizan_war_model(t, y, a_1, a_2, b_1, b_2, g_1, g_2):
    return np.array([-a_1*y[0] - b_2*y[1] + g_1, -a_2*y[1] - b_1*y[0]*y[1] + g_2])

fig = plt.figure()
ax = fig.add_subplot(111)
ax.grid()
ax.legend(['Армия 1', 'Партизаны'])

fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.grid()
plt.xlabel('Страна 1')


def partizan_war_plotter(a_1, a_2, b_1, b_2, g_1, g_2, max_time, start_value_N, start_value_M):
    ax.clear()
    ax1.clear()
    ax1.grid()
    ax.grid()
    plt.xlabel('Армия 1')
    plt.ylabel('Партизаны')

    solve = solve_ivp(lambda t, y: partizan_war_model(t, y, a_1, a_2, b_1, b_2, g_1, g_2),
                      (0,max_time), [start_value_N, start_value_M], method = 'RK23')
    
    l1, = ax.plot(solve['t'][(solve['y'][0]>-1) & (solve['y'][1]>-1)], solve['y'][0][(solve['y'][0]>-1) & (solve['y'][1]>-1)])
    l2, = ax.plot(solve['t'][(solve['y'][0]>-1) & (solve['y'][1]>-1)], solve['y'][1][(solve['y'][0]>-1) & (solve['y'][1]>-1)])
    ax1.plot(solve['y'][0][(solve['y'][0]>-1) & (solve['y'][1]>-1)], solve['y'][1][(solve['y'][0]>-1) & (solve['y'][1]>-1)])
    ax.legend((l1, l2), ('Армия 1', 'Партизаны'))
interact(partizan_war_plotter,
    a_1=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    a_2=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    b_1=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    b_2=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    g_1=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    g_2=widgets.FloatSlider(min=0, max=10, step=0.1, value=1),
    max_time=widgets.IntSlider(min=4, max=400, step=1, value=5),
    start_value_N=widgets.IntSlider(min=2, max=200, step=10, value=10,style = style),
    start_value_M=widgets.IntSlider(min=2, max=200, step=10, value=10,style = style)
);
```
