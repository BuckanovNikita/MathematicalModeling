```python
%matplotlib notebook
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
import ipywidgets as widgets
import scipy.integrate as integrate
import seaborn as sns
style = {'description_width': 'initial'}
```

```python
N = 10
x = 200*np.random.randn(N,N)
x = x + x.T
for i in range(N):
    x[i][i] = 0
    for j in range(i):
        x[i][j] = -x[i][j]
sns.heatmap(x)
X = np.sum(abs(x))
display(f'X: {X}')
s = np.sum(x, axis=1)
plt.figure()
plt.grid()
sns.barplot(list(range(N)),s)
S = np.sum(abs(s))
display(f'S: {S}')
```

```python
x_ = np.zeros(x.shape)
for i in range(N):
    for j in range(N):
        x_[i][j] = (s[i]*abs(s[j]) - s[j]*abs(s[i]))/S
plt.figure()
plt.grid()
sns.heatmap(x_)
plt.figure()
plt.grid()
sns.heatmap(abs(x_)<10)
X_ = np.sum(abs(x_))
display(f'X_: {X_}')
s_ = np.sum(x_, axis=1)
plt.figure()
plt.grid()
sns.barplot(list(range(N)),s_)
S_ = np.sum(abs(s_))
display(f'S_: {S_}')
```
